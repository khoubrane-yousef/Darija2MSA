```python
import pandas as pd
import re
```

```python
from datasets import load_dataset

dataset = load_dataset("M-A-D/DarijaBridge")
dataset = dataset['train'].to_pandas()
dataset
```

```python
dataset_translated = dataset[dataset.translated]
dataset_translated
```

```python
dataset_translated.translation.loc[1235090]
```

The translation column is not a clean English counterpart of the Darija sentence, it is a prompt that was given to LLMs by the creator of the dataset to generate the translation.

**Methodology**: "Data is collected through community contributions, ensuring a diverse representation of dialects and usage. Each sentence in Darija is paired with its English translation, reviewed and corrected by language experts and expert models (like GPT-4) for accuracy." (Source: https://huggingface.co/datasets/M-A-D/DarijaBridge)

The English translation is contained inside the prompt and there are three prompting patterns. Therefore, we need to clean the prompt according to these patterns to get our English translations.

```python
prompt_templates = list(dataset_translated['metadata'].apply(lambda x: x['template']).unique())
prompt_templates
```

```python
example_pattern1 = dataset_translated[dataset_translated.metadata.apply(lambda x: x['template'] == prompt_templates[0])].sample().translation.item()
example_pattern2 = dataset_translated[dataset_translated.metadata.apply(lambda x: x['template'] == prompt_templates[1])].sample().translation.item()
example_pattern3 = dataset_translated[dataset_translated.metadata.apply(lambda x: x['template'] == prompt_templates[2])].sample().translation.item()

print('Here is an example of the first pattern (command): ', example_pattern1)
print('\nHere is an example of the second pattern (question): ', example_pattern2)
print('\nHere is an example of the third pattern (continuation): ', example_pattern3)
```

```python
def preprocess_translation(text, pattern):
    if pattern == 0:
        return re.sub(re.escape(' Give me the same text in Moroccan Arabic.'), '', text)
    elif pattern == 1:
        return re.sub(re.escape('A text in English: ') + '|' + re.escape("\nWhat's the text in Moroccan Arabic?"), '', text)
    elif pattern == 2:
        return re.sub(re.escape(' The previous text is in English. Here is a translation to Moroccan Arabic:'), '', text)
```

```python
preprocess_translation(example_pattern3, 2)
```

```python
for pattern in range(len(prompt_templates)):
    mask = dataset_translated['metadata'].apply(lambda x: x.get('template') == prompt_templates[pattern])
    dataset_translated.loc[mask, 'translation'] = dataset_translated.loc[mask, 'translation'].apply(lambda x: preprocess_translation(x, pattern))
```

```python
DarijaEnglish_dataset = dataset_translated[['sentence', 'translation']].reset_index(drop=True).copy()
DarijaEnglish_dataset.rename(columns={'sentence': 'Darija', 'translation': 'English'}, inplace=True)
DarijaEnglish_dataset
```

```python
DarijaEnglish_dataset.to_csv('datasets/DarijaEnglish.csv', index=False)
```
